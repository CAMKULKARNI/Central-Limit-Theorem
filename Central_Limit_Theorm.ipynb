```python
import random
import matplotlib.pyplot as plt
```

```python
n = 10000
```

```python
def mu_sigma(array):
  mu = sum(array) / len(array)
  sigma = 0
  for i in array:
    sigma += ((i - mu) ** 2)

  sigma /= len(array)
  sigma = sigma ** 0.5
  return mu, sigma
```

```python
for sample_size in [10**i for i in range(5)]:
  means = []
  sds = []
  for _ in range(n):
    array = []
    for i in range(sample_size):
      array.append(random.uniform(0, 1))

    mu, sigma = mu_sigma(array)
    means.append(mu)
    sds.append(sigma)
  plt.hist(means, bins=100)
  plt.show()
  plt.hist(sds, bins=100)
  plt.show()
  print("Sample Size : ", sample_size)
```
